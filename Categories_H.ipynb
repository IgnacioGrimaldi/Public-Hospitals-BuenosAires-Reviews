```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

df = pd.read_csv("reviews_748_clasificadas.csv")
df["categorias_lista"] = df["categorias"].apply(
    lambda s: [c.strip() for c in str(s).split(";") if c.strip()] if pd.notna(s) else []
)

df["anio"] = pd.to_datetime(df["fecha_iso"]).dt.year

CATEGORIAS = [
    "delays",
    "inappropriate_interpersonal_behavior",
    "inefficient_medical_treatment",
    "bad_infrastructure",
    "incorrect_diagnoses",
    "other",
]

print(f"Cargadas {len(df)} reseñas")
print(f"Hospitales: {df['hospital'].unique().tolist()}")
print(f"Años: {sorted(df['anio'].unique().tolist())}")
df.head()
```

```python
total = len(df)
conteo = Counter(cat for lista in df["categorias_lista"] for cat in lista)

tabla = pd.DataFrame(
    [(cat, n, n/total*100) for cat, n in conteo.most_common()],
    columns=["categoria", "cantidad_filas", "porcentaje"]
)
tabla["porcentaje"] = tabla["porcentaje"].round(1)

print(f"Total de reseñas: {total}\n")
print(tabla.to_string(index=False))
```

```python
#Let's see how is this percentage but by hospital
```

```python
filas = []
for hosp in df["hospital"].unique():
    sub = df[df["hospital"] == hosp]
    total_hosp = len(sub)
    conteo = Counter(cat for lista in sub["categorias_lista"] for cat in lista)
    fila = {"hospital": hosp, "total_reviews": total_hosp}
    for cat in CATEGORIAS:
        fila[cat] = round(conteo.get(cat, 0) / total_hosp * 100, 1)
    filas.append(fila)

tabla_hosp = pd.DataFrame(filas).set_index("hospital")
print("Porcentaje de reseñas que menciona cada categoría, por hospital (%):\n")
print(tabla_hosp.to_string())
```

```python
tabla_hosp = pd.DataFrame(filas).set_index("hospital")
tabla_hosp
```

```python
total = len(df)
conteo = Counter(cat for lista in df["categorias_lista"] for cat in lista)

NOMBRES = {
    "delays": "Demoras",
    "inappropriate_interpersonal_behavior": "Mal trato",
    "inefficient_medical_treatment": "Tratamiento deficiente",
    "bad_infrastructure": "Infraestructura",
    "incorrect_diagnoses": "Diagnóstico erróneo",
    "other": "Otro",
}

filas = []
for cat, n in conteo.most_common():
    pct = n / total * 100
    filas.append({
        "categoria": NOMBRES.get(cat, cat),
        "cantidad": n,
        "porcentaje": round(pct, 1),
        "de_cada_10": round(pct / 10, 1)
    })

tabla_general = pd.DataFrame(filas)
tabla_general.to_csv("categorias_general.csv", index=False, encoding="utf-8-sig")
tabla_general
```
